## Simple example

```python
import pandas as pd
from ipyvizzu import Chart, Data, Config, Style

#Add Titanic data set to a data frame with a Count value added to it
data_frame = pd.read_csv(
    "https://raw.githubusercontent.com/vizzuhq/ipyvizzu/main/docs/examples/stories/titanic/titanic.csv"
)
data = Data()
data.add_data_frame(data_frame)

data_frame.head(4)
```

```python
chart = Chart(width="100%", height="360px")
chart.animate(data,
    Config(
        {
            "x": "Count",
            "y": "Sex",
            "label": "Count",
            "title": "Passengers of the Titanic",
        }
    )
)
```

```python
chart.animate(
    Config(
        {
            "x": ["Count", "Survived"],
            "color": "Survived"
        }
    )
)
```

```python
chart.animate(
    Config(
        {
            "x": "Count", 
            "y": ["Sex", "Survived"],
            #"geometry":"circle",
            "size":"Count",
            "coordSystem":"polar"
        }
    )
)
```

## Config
We have axes (x,y) and channels (color, label, size, lightness)

Axes can have any number of categorical and one numerical series on them.


Other parameters include:           
- coordSystem (*cartesian*, polar)
- geometry (*rectangle*, circle, area, line)
- sort (*none*, byValue)
- align (*none*, min, max, center **stretch**)
- title (string)
- split (true, *false*)

*default values*
